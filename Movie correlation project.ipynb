```python
# import lbraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6)

#read in data

df= pd.read_csv(r'C:\Users\USER\Desktop\movies.csv.')
```

```python
#let look at the data 
df.head()
```

```python
# checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format (col, pct_missing))
```

```python
#data types for our columns
df.dtypes
```

```python
# replace null values for budget and gross colume to zero

cols = ['budget', 'gross']

df[cols] = (
    df[cols]
    .replace([np.inf, -np.inf], np.nan)
    .fillna(0)
    .astype(int)
)
```

```python
# change data type for colume

df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending = False)
df.head()
```

```python
pd.set_option('display.max_rows', None)
df.head()
```

```python
# drop any duplicate

df['company'].drop_duplicates().sort_values(ascending = False)

```

```python
# scatterd plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('budget vs gross earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
df.head()
```

```python
# plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
plt.show()
```

```python
#let start looking at correlation

correlation_matric = df.corr(numeric_only=True) #pearson, kendall, spearman
correlation_matric
```

```python
df.corr(numeric_only =True, method='pearson')
sns.heatmap(correlation_matric, annot = True)
plt.title('correlation matric for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] =  df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()     
```

```python
correlation_matric = df_numerized.corr(method= 'pearson')
sns.heatmap(correlation_matric, annot = True)
plt.title('correlation matric for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_correlation=sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```
